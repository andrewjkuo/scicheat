```python
import pandas as pd
from scicheat.dataloader import PrimaryAnalysis
```

```python
data_url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/statlog/heart/heart.dat'
names = ['age', 'sex', 'pain_type', 'rest_bp', 'serum_chol', 'fast_bl_sugar',
         'rest_ec_val', 'max_hr', 'exercise_angina', 'oldpeak', 'peak_slope',
         'num_of_maj_vessels', 'thal', 'heart_disease_presence']

df = pd.read_csv(data_url, names=names, sep=' ')
df.head()
```

```python
pa = PrimaryAnalysis(df, 'heart_disease_presence')
pa.run_all()
```
